```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.special as sps
```

# 1

```python
n, p = 30, .2  
loi_binomiale = np.random.binomial(n, p, 10000)
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax = sns.distplot(loi_binomiale, bins=11, label='resultats de la simulation')
ax.set_xlabel("Nombre de s",fontsize=16)
ax.set_ylabel("Frequence",fontsize=16)
```

# 2

```python
mu, sigma = 3, 4 # mean and standard deviation
loi_normale = np.random.normal(mu, sigma, 10000)
```

```python
count, bins, ignored = plt.hist(loi_normale, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
linewidth=2, color='r')
plt.show()
```

# 3

```python
shape, scale = 10, 5  
loi_gamma = np.random.gamma(shape, scale, 10000)
```

```python
count, bins, ignored = plt.hist(loi_gamma, 50, density=True)
y = bins**(shape-1)*(np.exp(-bins/scale) /(sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r')
plt.show()
```
